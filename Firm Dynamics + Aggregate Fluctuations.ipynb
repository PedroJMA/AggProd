# Firm dynamics and aggregate fluctuations

## Overview

## References 

* Clementi - Palazzo (2016) Entry, exit, firm dynamics, and aggregate fluctuations

## The Economy

Time is discrete and is indexed by $t = 1, 2, \ldots$.
The horizon is infinite.

### Firms

At time $t$, a positive mass of price-taking firms produce an homogeneous good by means of the production function
$$
y_t = z_t s_t (k_t^\alpha l_t^{1-\alpha})^\theta
$$
where $\alpha, \theta \in (0,1)$.
With $k_t$ we denote physical capital, $l_t$ is labor, and $z_t$ and $s_t$ are aggregate and idiosyncratic random disturbances, respectively.

The common component of productivity $z_t$ is driven by the stochastic process
$$
\log z_{t+1} = \rho_z \log z_t + \sigma_z \varepsilon_{z,t+1}
$$
where $\varepsilon_{z,t+1} \sim N(0,1)$ for all $t \geq 0$.

The dynamics of the idiosyncratic component $s_t$ is described by
$$
\log s_{t+1} = \rho_s \log s_t + \sigma_s \varepsilon_{s,t+1}
$$
with $\varepsilon_{s,t+1} \sim N(0,1)$ for all $t \geq 0$.

The conditional distribution of $s_{t+1}$ will be denoted as $H(s_{t+1}|s_t)$.

Firms hire labor services on the spot market at the wage rate $w_t$ and discount future profits by means of the time-invariant factor $\frac{1}{R}$, $R > 1$.

Adjusting the capital stock by $x$ bears a cost $g(x,k)$.

Capital depreciates at the rate $\delta \in (0,1)$.

We assume that the demand for the firm's output and the supply of capita are infinitely elastic and normalize their prices at one.

The supply of labor is given by the function
$$
L_s(w) = w^\gamma \text{, with } \gamma > 0
$$

Operating firms incur a cost $c_f > 0$, drawn from the common time-invariant distribution $G$.
Firms that quit producing cannot reenter the market at a later stage and recoup the undepreciated portion of their capital stock, net of the adjustment cost of driving it to zero.


### Timing

Every period there is a constant mas $M > 0$ of prospective entrants, each of which receives a signal $q$ about her productivity, with $q \sim Q(q)$.

Conditional on entry, the distribution of the idiosyncratic shock in the first period of operation is $H(s'|q)$, strictly decreasing in $q$.
Entrepreneurs that decide to enter the industry pay an entry cost $c_e \geq 0$.

At all $t \geq 0$, the distribution of operating firms over the two dimensions of heterogeneity is denoted by $\Gamma_t(k,s)$.

Finally, let $\lambda_t \in \Lambda$ denote the vector of aggregate state variables and $J(\lambda_{t+1}|\lambda_t)$ its transition operator.


$$
\lambda_t = \left\{ \Gamma_t, z_t \right\}
$$


## The Incumbent's Optimization Program

Given the aggregate state $\lambda$, capital in place $k$, and idiosyncratic productivity $s$, the employment choice is the solution to the following static problem:
$$
\pi(\lambda, k, s) = \max_l sz[k^\alpha l^{1-\alpha}]^\theta - wl
$$

Upon exit, a firm obtains a value equal to the undepreciated portion of its capital $k$, net of the adjustment cost it incurs in order to dismantle it, i.e.
$$
V_x(k) = k(1-\delta) - g[-k(1-\delta), k]
$$

Then, the start-of-period value of an incumbent firm is dictated by the function $V(\lambda, k,s)$ which solves the following functional equation:
$$
V(\lambda, k, s) = \pi(\lambda, k, s) + \int \max \left\{ V_x(k), \tilde{V}(\lambda, k, s) - c_f \right\} dG(c_f)
$$

\begin{align*}
\tilde{V}(\lambda, k, s) &= \max_x - x - g(x,k) + \frac{1}{R} \int_{\Lambda} \int_{\Re} V(\lambda^\prime, k^\prime, s^\prime) dH(s^\prime | s) dJ(\lambda^\prime | \lambda) \\
&\quad \text{s.t. } k^\prime = k(1-\delta) + x
\end{align*}


## Entry 

For an aggregate state $\lambda$, the value of a prospective entrant that obtains a signal $q$ is
$$
V_e(\lambda, q) = \max_{k^\prime} - k^\prime + \frac{1}{R} \int_{\Lambda} \int_{\Re} V(\lambda^\prime, k^\prime, s^\prime) dH(s^\prime | s) dJ(\lambda^\prime | \lambda)
$$
She will invest and start operating if and only if
$$
V_e(\lambda, q) \geq c_e
$$

## Recursive Competitive Equilibirum 



For given $\Gamma_0$, a recursive competitive equilibrium consist of

* (i) value functions $V(\lambda, k, s)$, $\tilde{V}(\lambda, k, s)$, and $V_e(\lambda, q)$;
* (ii) policy functions $x(\lambda, k, s)$, $l(\lambda, k, s)$, $k^\prime(\lambda, q)$; and
* (iii) bounded sequences of wages
$\left\{ w_t \right\}_{t=0}^\infty$,
incumbents' measures $\left\{ \Gamma_t \right\}_{t=1}^\infty$, and
entrants' measures
$\left\{ \mathcal{E}_t \right\}_{t=0}^\infty$
such that, for all $t \geq 0$:


1. $V(\lambda, k, s)$, $\tilde{V}(\lambda, k, s)$, $x(\lambda, k, s)$, and $l(\lambda, k, s)$ solve the incumbent's problem;
2. $V_e(\lambda, k, s)$ and $k^\prime(\lambda, q)$ solve the entrant's problem;
3. the labor market clears:
$$
\int l(\lambda_t, k, s) d\Gamma_t(k,s) = L_s(w_t) \quad \forall t \geq 0
$$
4.  for all Borel sets $\mathcal{S}\times \mathcal{K} \in \Re^{+} \times \Re^{+}$ and $\forall t \geq 0$,
$$
\mathcal{E}_{t+1}\left( \mathcal{S}\times \mathcal{K} \right) = M \int_{\mathcal{S}} \int_{\mathcal{B}_e(\mathcal{K}, \lambda_t)} dQ(q) dH(s^\prime|q)
$$
where
$$
\mathcal{B}_e(\mathcal{K}, \lambda_t) = \left\{ q \text{ s.t. } k^\prime(\lambda_t, q) \in \mathcal{K} \text{ and } V_e(\lambda_t, q) \geq c_e \right\}
$$
5.  for all Borel sets $\mathcal{S}\times \mathcal{K} \in \Re^{+} \times \Re^{+}$ and $\forall t \geq 0$,
$$
\Gamma_{t+1}(\mathcal{S}\times\mathcal{K}) = \int_{\mathcal{S}} \int \int_{\mathcal{B}(\mathcal{K}, \lambda_t, c_f)} d\Gamma_t(k,s) dG(c_f) dH(s^\prime | s) + \mathcal{E}_{t+1}(\mathcal{S}\times\mathcal{K})
$$
where
$$
\mathcal{B}(\mathcal{K}, \lambda_t, c_f) = \left\{ (k,s) \text{ s.t. } \tilde{V}(\lambda_t, k, s) - c_f \geq V_x(k) \text{ and } k(1-\delta) + x(\lambda_t, k, s) \in \mathcal{K} \right\}
$$

## Functional Forms 

Investment adjustment costs are the sum of a fixed portion and of a convex portion:

$$
g(x,k) = \chi(x) c_0k + c_1 \left( \frac{x}{k} \right)^2 k, \quad c_0, c_1 \geq 0
$$
where $\chi(x) = 0$ for $x=0$ and $\chi(x)=1$ otherwise.
Notice that the fixed portion is scaled by the level of capital in place and is paid if and only if gross investment is different from zero.

The distribution of signals for the entrants is Pareto.
We posit that $q \geq \underline{q} \geq 0$
and that
$$
Q(q) = 1 - (\underline{q}/q)^{\xi}, \quad \xi > 1
$$
The realization of the idiosyncratic shock in the first period of operation follows the process
$$
\log(s) = \rho_s \log(q) + \sigma_s \eta, \text{ where } \eta \sim N(0,1)
$$

Finally, we assume that the distribution of the operating cost $G$ is log-normal with parameters $\mu_{c_f}$ and $\sigma_{c_f}$.


## Entry, Investment, and Exit

Since an incumbent's value $V(k,s)$ is weakly increasing in the idiosyncratic productivity shock $s$ and the conditional distribution $H(s^\prime|q)$ is decreasing in the signal $q$, the value of entering $V_e(q)$ is strictly increasing function of the signal.
In turn, this means that there will be a threshold for $q$, call it $q^*$, such that prospective entrants will enter if and only if they receive a draw greater than or equal to $q^*$.

The mass of actual entrants with productivity less than or equal to any $\bar{s} \geq 0$ will be
$$
M \int_{q^*}^\infty d H(s | q) dQ(q)
$$

Let $k^\prime(q)$ denote the optimal entrants' capital choice conditional on having received a signal $q$.
For every $\bar{k} \geq k^\prime(q^*)$ and for all $\bar{s} \geq 0$, the portion of entrants characterized by pairs $(s,k)$ such that $s \leq \bar{s}$ and $k \leq \bar{k}$ will be
$$
M \int_0^{\bar{s}} \int_{q^*}^{q^{-1}(\bar{k})} dQ(q) dH(s|q)
$$
where $q^{-1}$ is the inverse of the mapping between signal and capital choice.


### Notes

* Prospective entrants are heterogeneous. Some obtain a greater signal than others and therefore face better short-term prospects. As a consequence, they start with greater capital and are likely to hire more employees.
* Investment is modeled as in the standard neoclassical framework with nonconvex adjustment costs.
    * The fixed cost of investment gives rise to policy functions of the $S-s$ type, which in turn are responsible for the periods of inaction that are characteristic of all firm- and plant-level datasets.
* Firms exit whenever they draw an operating cost $c_f$ such that the value of exiting $V_x(k)$ is greater than the value of continuing operations, i.e., $\tilde{V}(k,s) - c_f$.
    * For a given capital, the probability of exiting is decreasing in the idiosyncratic productivity.This is not surprising, given that the value of exiting does not depend on $s$, while the continuation payoff is obviously greater, the larger is productivity.Survival chances also improve with capital, for a given level of productivity.
* all firms survive with positive probability and survival is smoothly increasing in both capital and productivity.
    * If we let the distribution of the operating cost collapse to a singleton, the exit policy would be characterized by a decreasing schedule, call it $\underline{s}(k)$, such that a firm equipped with capital $k$ will exit if and only if its productivity is lower than $\underline{s}(k)$.
